```python
import numpy as np
import pandas as pd
np.random.seed(50)

data = {
    "x" : np.arange(1,14),
    "y" : np.random.randint(9,99,size=13)
}
df = pd.DataFrame(data)
print(df)

import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.plot(df["x"],df["y"],marker='.',color = "purple",linestyle= '--')

plt.xlabel("days")
plt.ylabel("Time")
```

```python
import numpy as np
import pandas as pd

data = {
    "x": np.arange(0, 11),
    "y": np.arange(0, 11)
}


df = pd.DataFrame(data)
print(df)

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(df["x"], df["y"], marker='o', linestyle='--')
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Line Graph")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(size=200)

plt.figure(figsize=(10, 5))
plt.hist(data, bins=20, edgecolor='black')

plt.xlabel("a")
plt.ylabel("b")
plt.title("Histogram")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 11)
y = x**2

plt.figure(figsize=(10, 8))


plt.subplot(2, 2, 1)
plt.plot(x, y)
plt.title("Line Plot")

plt.subplot(2, 2, 2)
plt.bar(x, y)
plt.title("Bar Plot")

plt.subplot(2, 2, 3)
plt.scatter(x, y)
plt.title("Scatter Plot")

plt.subplot(2, 2, 4)
plt.hist(y, bins=5)
plt.title("Histogram")

plt.tight_layout()
plt.show()
```
